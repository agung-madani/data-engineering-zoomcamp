```python
import pandas as pd
from sqlalchemy import create_engine
from tqdm.auto import tqdm
```

```python
prefix = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/"
file_name = "yellow_tripdata_2021-01.csv.gz"
csv_url = prefix + file_name
```

```python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]
```

```python
df = pd.read_csv(
    csv_url,
    nrows=100,
    dtype=dtype,
    parse_dates=parse_dates
)

df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
engine = create_engine(
    "postgresql://postgres:postgres@localhost:5433/ny_taxi?sslmode=disable"
)
```

```python
print(
    pd.io.sql.get_schema(
        df,
        name="yellow_taxi_data",
        con=engine
    )
)
```

```python
df.head(0).to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="replace",
    index=False
)

print("Table created")
```

```python
df_iter = pd.read_csv(
    csv_url,
    iterator=True,
    chunksize=100000,
    dtype=dtype,
    parse_dates=parse_dates
)
```

```python
df_first = next(df_iter)

df_first.to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="append",
    index=False
)

print("Inserted first chunk:", len(df_first))
```

```python
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append",
        index=False
    )
```
